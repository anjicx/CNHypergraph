For creating hyperedges we have to find uniques stays(1 stay is repeated because of different diagnosis in one stay). Then we need a dictionary visit1:hyperedge_index where visit_11:0 for instance.

```python

# this is hyperedge_id:visit
stay_hyperedge={}#define it is dict
for i in range(len(unique_stays)):#this is same because the length is the length of stays
  stay_hyperedge[i]=unique_stays[i]

# fvisit_id:hyperedge_id
stay_to_hyperedge = {}
for i in range(len(unique_stays)):
  stay_id=unique_stays[i]
  stay_to_hyperedge[stay_id] = i #Or   stay_hyperedge[unique_stays[i]]=i

#IMPORTANT!
#THIRD WAY TO SOLVE THIS
#range(len(x)) gives index enumarate(x)  gives index and value so:
stay_to_hyperedge = {}
for i, stay_id in enumerate(unique_stays):#this is same because the length is the length of stays
  stay_to_hyperedge[stay_id]=i
```
